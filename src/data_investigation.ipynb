```r
library(lme4)
library(lmerTest)
library(dplyr)
```

```r
run_glmer <- function(data_df){
    model = glmer(TIME_POINT ~ chem_abundance + sex + (1|ANIMAL_ID), data = data_df, family = binomial)
    pval_list = coef(summary(model))[,4]
    pval = pval_list[2]

    if (pval == 0){
        pval <- "error"
    }
    return (pval)
}
```

```r
run_glm <- function(data_df){
    model = glm(TIME_POINT ~ chem_abundance, data = data_df, family = binomial)
    pval_list = coef(summary(model))[,4]
    pval = pval_list[2]

    if (pval == 0){
        pval <- "error"
    }
    return (pval)
}
```

```r
run_wttest <- function(before_list, after_list){
    
    ttest = t.test(before_list, after_list)
    pval = ttest[["p.value"]]

    return (pval)
}
```

```r
data_file = '../data/fecal_mouse_metabolites_scaled.csv'
data_df = read.csv(data_file, row.names=1)
data_df = t(data_df)

col_length = ncol(data_df)

feature_list = colnames(data_df)
feature_size = length(feature_list) 

```

```r
#mixed-effects
output_txt = '../data/fecal_mouse_metabolites_scaled.glmer.txt' 

if (file.exists(output_txt)){
    file.remove(output_txt)
}

output_string <- paste("chemID\tpval\n", sep="")
cat(output_string, file=output_txt, append=TRUE)

for (i in 4:col_length){
   
    feature = feature_list[i]

    subset_df = subset(data_df, select = c(1:3,i))
    colnames(subset_df)[4] <- "chem_abundance"
    subset_df = as.data.frame(subset_df)
    pval <- tryCatch(run_glmer(subset_df), error = function(e){ return ("error")})

    output_string <- paste(feature,"\t",pval,"\n", sep="")
    cat(output_string, file=output_txt,append=TRUE)
       
}
```

```r
#GLM
output_txt = '../data/fecal_mouse_metabolites_scaled.glm.txt' 

if (file.exists(output_txt)){
    file.remove(output_txt)
}

output_string <- paste("chemID\tpval\n", sep="")
cat(output_string, file=output_txt, append=TRUE)

for (i in 4:col_length){
   
    feature = feature_list[i]

    subset_df = subset(data_df, select = c(1:3,i))
    colnames(subset_df)[4] <- "chem_abundance"
    subset_df = as.data.frame(subset_df)
    pval <- tryCatch(run_glm(subset_df), error = function(e){ return ("error")})

    output_string <- paste(feature,"\t",pval,"\n", sep="")
    cat(output_string, file=output_txt,append=TRUE)
       
}
```

```r
#welch t-test
output_txt = '../data/fecal_mouse_metabolites_scaled.wttest.txt' 

if (file.exists(output_txt)){
    file.remove(output_txt)
}

output_string <- paste("chemID\tpval\n", sep="")
cat(output_string, file=output_txt, append=TRUE)

for (i in 4:col_length){
   
    feature = feature_list[i]

    subset_df = subset(data_df, select = c(1:3,i))
    colnames(subset_df)[4] <- "chem_abundance"
    subset_df = as.data.frame(subset_df)
    
    subset_df %>% filter(TIME_POINT == 0, )
    before_list = filter(subset_df, TIME_POINT == 0)$chem_abundance
    after_list = filter(subset_df, TIME_POINT == 1)$chem_abundance

    
    pval <- tryCatch(run_wttest(before_list,after_list) ,error = function(e){ return ("error")})
    output_string <- paste(feature,"\t",pval,"\n", sep="")
    cat(output_string, file=output_txt,append=TRUE)
#     break
}
```
